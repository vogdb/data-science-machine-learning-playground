```python
replace_dict = {'No': False, 'Yes': True}
pd_data_frame['International plan'] = pd_data_frame['International plan'].map(replace_dict)
pd_data_frame.head()
```

```python
pd_data_frame = pd_data_frame.replace({'Voice mail plan': replace_dict})
pd_data_frame.head()
```

## Grouping

```python
columns_to_show = ['Total day minutes', 'Total eve minutes', 
                   'Total night minutes']
pd_data_frame.groupby(['Churn'])[columns_to_show].describe(percentiles=[])
```

```python
pd_data_frame.groupby(['Churn'])[columns_to_show].agg([np.mean, np.std, np.min, np.max])
```

## Summary tables

```python
pd.crosstab(pd_data_frame['Churn'], pd_data_frame['International plan'])
```

```python
pd.crosstab(pd_data_frame['Churn'], pd_data_frame['Voice mail plan'], normalize=True)
```

```python
pd_data_frame.pivot_table(['Total day calls', 'Total eve calls', 'Total night calls'], ['Area code'], aggfunc=('mean', 'std'))
```

## Dataframe transformations

```python
total_calls = pd_data_frame['Total day calls'] + \
                pd_data_frame['Total eve calls'] + \
                pd_data_frame['Total night calls'] + \
                pd_data_frame['Total intl calls']
pd_data_frame.insert(
    loc=len(pd_data_frame.columns),
    column='Total calls',
    value=total_calls
)
```

```python
pd_data_frame.head()
```

```python
pd_data_frame['Total charge'] = pd_data_frame['Total day charge'] + \
                                pd_data_frame['Total eve charge'] + \
                                pd_data_frame['Total night charge'] + \
                                pd_data_frame['Total intl charge']
pd_data_frame.head()
```

## Predicting telecom churn

```python
pd.crosstab(pd_data_frame['Churn'], pd_data_frame['International plan'], margins=True)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt 
import seaborn as sns

plt.rcParams['figure.figsize'] = (8, 6)
```

```python
sns.countplot(x='International plan', hue='Churn', data=pd_data_frame)
```

```python
pd.crosstab(pd_data_frame['Churn'], pd_data_frame['Customer service calls'], margins=True)
```

```python
sns.countplot(x='Customer service calls', hue='Churn', data=pd_data_frame)
```

```python
pd_data_frame['Many service calls'] = (pd_data_frame['Customer service calls'] > 3).astype('int')
```

```python
pd.crosstab(pd_data_frame['Churn'], pd_data_frame['Many service calls'], margins=True)
```

```python
sns.countplot(x='Many service calls', hue='Churn', data=pd_data_frame)
```

```python
pd.crosstab(pd_data_frame['Many service calls'] & pd_data_frame['International plan'], pd_data_frame['Churn'])
```

```python
pd.crosstab([pd_data_frame['Many service calls'], pd_data_frame['International plan']], pd_data_frame['Churn'])
```
